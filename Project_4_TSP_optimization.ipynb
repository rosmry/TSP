The travelling salesman problem (also called the travelling salesperson problem or TSP) is a classic optimization problem that seeks to find the shortest possible route for a salesman to visit a set of cities and return to the starting point, it asks the following question: "Given a list of cities and the distances between each pair of cities, what is the shortest possible route that visits each city exactly once and returns to the origin city?" It can be solved using different methods, here we consider using TSP library in python.

Dataset :
We arev given with a disatnce matrix as our dataset along with the name of the cities to be travelled. we consider the Bengaluru  as our starting point and the final destination should be the origin city itself.In between the there are 13 other cities whose distance with each other is given in KMs.

Constraints:
In this optimization problem we are given with certain constraints they are, the starting and returning point should be the warehouse. Each city should be travelled only once. Also only one set of route is needed.

```python
import pandas as pd
import numpy as np
from python_tsp.exact import solve_tsp_dynamic_programming
import io
from google.colab import files
uploaded = files.upload()
```

```python
## reading dataset
data = pd.read_excel(io.BytesIO(uploaded['TSP_dataset (1).xlsx']))
data
```

```python
df = data.set_index('location', inplace=True)
```

```python
distance_matrix = df.values
distance_matrix
```

```python
optimal_route, distance = solve_tsp_dynamic_programming(distance_matrix)
```

```python
print(optimal_route)
```

```python
print("The minimum distance taken for covering all cities: {:.1f} KM".format(distance))
```

```python
Locations = df.index.tolist()
```

```python
from geopy.geocoders import Nominatim
import folium

# Assuming you have a list called 'locations' containing the indices of the cities
# Assuming you have the optimal route order stored in a list called 'optimal_route'
# Assuming you have a DataFrame called 'df' containing the distance matrix and city names

# Create a map
m = folium.Map(location=[0, 0], zoom_start=2)

# Create a geocoder object
geolocator = Nominatim(user_agent="my_app")

# Plot the optimal route on the map
for i in range(len(optimal_route) - 1):
    index1 = optimal_route[i]
    index2 = optimal_route[i+1]

    # Get the city names from the DataFrame using the indices
    city1 = df.loc[locations[index1], 'location']
    city2 = df.loc[locations[index2], 'location']

    # Get the coordinates of the cities
    location1 = geolocator.geocode(city1)
    location2 = geolocator.geocode(city2)

    # Check if coordinates are found
    if location1 is not None and location2 is not None:
        # Plot a line between the cities
        folium.PolyLine(locations=[(location1.latitude, location1.longitude), (location2.latitude, location2.longitude)], color='red').add_to(m)

# Save the map as an HTML file
m.save('optimal_route_map.html')

# Display the map
m
```

The ouput of the TSP gives us the order in which the saleman need to visit the cities.

The order in which the salesman need to travel is : 0->7->12->13->9->11->2->3->6->8->4->5->1->10->0.  

The order of the city: Banglore-> Mysore-> Shivamogga-> Hubballi-> Mangluru-> Kozhikode-> Kochi-> Trivandrum-> Madurai-> Coimbatore-> Salem-> Pondicherry-> Chennai-> Vellore-> Banglore.


The distance that the salesman need to travel is 2992.6 km to cover all the cities and reach back the origin city.
